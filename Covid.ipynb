```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
españa = pd.read_csv('Spain.csv')

año = np.array(españa['Year'])
muertes = np.array(españa['Value'])

año=año[::-1]
muertes=muertes[::-1]
n=len(muertes)
x = np.linspace(1,n,1) 
i_noCovid = np.where(año<2019)
i_Covid = np.where(año>2018)
plt.figure()
plt.plot(muertes[i_noCovid])
plt.plot(muertes[i_Covid])

diferencia_real = np.mean(muertes[i_noCovid])-np.mean(muertes[i_Covid])
```

```python
n_NC=len(muertes[i_noCovid])
n_iteraciones = 500000
diferencias = np.zeros(n_iteraciones)

for i in range(n_iteraciones):
    np.random.shuffle(muertes)
    muertes_pandemia = muertes[:n_NC]
    muertes_sinPandemia = muertes[n_NC:]
    media_pandemia = np.mean(muertes_pandemia)
    media_sinPandemia = np.mean(muertes_sinPandemia)
    diferencias[i] = media_pandemia-media_sinPandemia
```

```python
_=plt.hist(diferencias, bins = 20)
plt.vlines(diferencia_real,0 , 8400, color="red")
```
